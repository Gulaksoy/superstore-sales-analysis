```python
df = pd.read_csv("/kaggle/input/datasets/bravehart101/sample-supermarket-dataset/SampleSuperstore.csv")
df.head()
```

```python
df.info()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
category_sales = df.groupby("Category")["Sales"].sum().reset_index()
category_sales
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10,6))
plt.bar(category_sales['Category'], category_sales["Sales"], color='skyblue')
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.ylim(0, 300000)
plt.show()
```

```python
category_profit = df.groupby("Category")["Profit"].sum().reset_index()
category_profit
```

```python
plt.figure(figsize=(10,6))
plt.bar(category_profit['Category'], category_profit['Profit'], color='salmon')
plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.ylim(-70000, 10000)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(data=category_profit, x="Category", y="Profit", palette="Reds_d")
plt.title("Profit by Category")
plt.show()
=plt.ylim(-70000, 10000)

```

Executive Insights (Business Summary)
1) Category Insights
Technology delivers the highest profitability, indicating strong margins and efficient pricing.

Furniture generates high sales volume but operates at a loss, suggesting issues in cost structure or discounting strategy.

Office Supplies performs steadily with moderate sales and low margins.

2) Profitability Insights
Overall profitability is negative, signaling that the current pricing and discount strategy is not sustainable.

Losses are primarily driven by the Furniture category, which requires immediate review.

3) Regional Insights
The East region leads in total sales, representing the strongest market opportunity.

The South region underperforms and may benefit from targeted marketing or customer acquisition initiatives.

Central and West regions show stable but moderate performance.

4) Segment Insights
Consumer segment is the primary revenue driver with the highest sales volume.

Corporate segment shows consistent but limited growth potential.

Home Office segment remains the smallest and could be improved through tailored promotions.

5) Actionable Recommendations
Conduct a cost and pricing review for the Furniture category to address negative margins.

Expand the premium product strategy within the Technology category to leverage strong profitability.

Implement localized campaigns in the South region to improve market penetration.

Increase personalized promotions for the Consumer segment to maximize revenue.

Optimize discount levels across all categories to improve overall profitability.
